# Import and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

# Load cleaned data
fraud_df = pd.read_csv('../data/cleaned/fraud_data_cleaned.csv')
creditcard_df = pd.read_csv('../data/cleaned/creditcard_cleaned.csv')
```

# EDA on E-commerce Fraud Data (fraud_df)

```python
# Distribution of purchase_value
plt.figure(figsize=(8,4))
sns.histplot(fraud_df['purchase_value'], bins=30, kde=True)
plt.title('Distribution of Purchase Value')
plt.show()

# Distribution of age
plt.figure(figsize=(8,4))
sns.histplot(fraud_df['age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()

# Source distribution
plt.figure(figsize=(6,4))
sns.countplot(data=fraud_df, x='source', order=fraud_df['source'].value_counts().index)
plt.title('Source Distribution')
plt.show()

# Browser distribution
plt.figure(figsize=(6,4))
sns.countplot(data=fraud_df, x='browser', order=fraud_df['browser'].value_counts().index)
plt.title('Browser Distribution')
plt.show()

# Sex distribution
plt.figure(figsize=(4,4))
sns.countplot(data=fraud_df, x='sex')
plt.title('Sex Distribution')
plt.show()

# Fraud class distribution
plt.figure(figsize=(4,4))
sns.countplot(data=fraud_df, x='class')
plt.title('Fraud Class Distribution')
plt.show()

print('Fraud Rate:')
print(fraud_df['class'].value_counts(normalize=True))
```

# BIvariate Analysis

```python
# Purchase value by fraud class
plt.figure(figsize=(6,4))
sns.boxplot(data=fraud_df, x='class', y='purchase_value')
plt.title('Purchase Value by Fraud Class')
plt.show()

# Age by fraud class
plt.figure(figsize=(6,4))
sns.boxplot(data=fraud_df, x='class', y='age')
plt.title('Age by Fraud Class')
plt.show()

# Source vs fraud class
plt.figure(figsize=(6,4))
sns.countplot(data=fraud_df, x='source', hue='class')
plt.title('Source vs Fraud Class')
plt.show()

# Browser vs fraud class
plt.figure(figsize=(6,4))
sns.countplot(data=fraud_df, x='browser', hue='class')
plt.title('Browser vs Fraud Class')
plt.show()

# Sex vs fraud class
plt.figure(figsize=(4,4))
sns.countplot(data=fraud_df, x='sex', hue='class')
plt.title('Sex vs Fraud Class')
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(fraud_df[['purchase_value', 'age', 'class']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (E-commerce Data)')
plt.show()
```

# Time-based Features

```python
# Convert to datetime if not already
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])

fraud_df['purchase_hour'] = fraud_df['purchase_time'].dt.hour
fraud_df['purchase_dayofweek'] = fraud_df['purchase_time'].dt.dayofweek

plt.figure(figsize=(8,4))
sns.countplot(data=fraud_df, x='purchase_hour', hue='class')
plt.title('Purchase Hour vs Fraud Class')
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(data=fraud_df, x='purchase_dayofweek', hue='class')
plt.title('Purchase Day of Week vs Fraud Class')
plt.show()
```

# EDA on Credit Card Fraud Data (creditcard_df)

```python
# Distribution of transaction amount
plt.figure(figsize=(8,4))
sns.histplot(creditcard_df['Amount'], bins=30, kde=True)
plt.title('Distribution of Transaction Amount')
plt.show()

# Fraud class distribution
plt.figure(figsize=(4,4))
sns.countplot(data=creditcard_df, x='Class')
plt.title('Fraud Class Distribution (Credit Card)')
plt.show()

print('Fraud Rate (Credit Card):')
print(creditcard_df['Class'].value_counts(normalize=True))
```

# Bivariate Analysis

```python
# Transaction amount by fraud class
plt.figure(figsize=(6,4))
sns.boxplot(data=creditcard_df, x='Class', y='Amount')
plt.title('Transaction Amount by Fraud Class')
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(12,8))
corr = creditcard_df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap (Credit Card Data)')
plt.show()
```
